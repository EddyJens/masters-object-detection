```python
data_list_file_path = f'/data/output/mixed_data/mixed_test.json'
# data_list_file_path = f'/data/output/mixed_data/mixed_train_val0.json'
f = open(data_list_file_path)
data = json.load(f)

luna = []
luna_ares = []
hc = []
hc_ares = []
msd = []
msd_ares = []

# for content in tqdm(data['training']):
for content in tqdm(data['test']):

    img, origin, spacing = load_itk(content['image'])
    origin = np.array(list(reversed(origin)))
    spacing = np.array(list(reversed(spacing)))

    for true_box in content['box']:

        world_coord_cent = np.array((true_box[0], true_box[1], true_box[2]))
        world_coord_diam = np.array((true_box[3], true_box[4], true_box[5]))

        voxel_min, voxel_max = world_to_pixel(
            origin, spacing, world_coord_cent, world_coord_diam
        )

        a1 = voxel_max[0] - voxel_min[0]
        a2 = voxel_max[1] - voxel_min[1]
        a3 = voxel_max[2] - voxel_min[2]

        if 'LUNA' in content['image']:
            luna.append(a1 * a2 * a3)
            luna_ares.append(a1)
            luna_ares.append(a2)
            luna_ares.append(a3)

        if 'MSD' in content['image']:
            msd.append(a1 * a2 * a3)
            msd_ares.append(a1)
            msd_ares.append(a2)
            msd_ares.append(a3)

        if 'HC' in content['image']:
            hc.append(a1 * a2 * a3)
            hc_ares.append(a1)
            hc_ares.append(a2)
            hc_ares.append(a3)
```

```python
plt.hist(luna, 10, range=[0, 1e6])
plt.show()
```
